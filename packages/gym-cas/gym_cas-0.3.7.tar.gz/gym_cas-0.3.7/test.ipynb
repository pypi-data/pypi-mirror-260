```python
from symfit import parameters, variables, Fit, Model
from gym_cas import *
import numpy as np
from sympy import latex

V = [7.9,9.5,10.7,13.3]
P = [0.05372,0.1216,0.24289,0.32319]

xdata = np.array(V)
ydata = np.array(P)
yerr = np.array([0.1, 0.1, 0.1, 0.1])


f, p1 = regression_power(V,P,return_plot=True,show=False)
# print(dir(p1))
# print(dir(p1.series[1]))
# print(p1.series[1].expr)
# print(p1.legend)
# print(p1._use_latex)
# p1.series[1].label = p1.series[1].expr
# p1.series[1]._latex_label = f"${latex(p1.series[1].expr)}$"
p1.show()

a, b = parameters('a, b')
x, y = variables('x, y')
model = Model({y: a*x**3})

fit = Fit(model, x=xdata, y=ydata, sigma_y=yerr)
fit_result = fit.execute()
fit_result.value(a)
```
